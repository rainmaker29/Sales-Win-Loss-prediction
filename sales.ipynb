## We have several columns in the dataframe whose values are all strings. Our processing models are operate on integers.

### So, we need to encode these string values into integer values using the scikit-learn's LabelEncoder() methods


```python
#Let's first check out the unique values of all string type columns
```

```python
print("Supplies subgroup :",sales_data['Supplies Subgroup'].unique())
print("Region :",sales_data['Region'].unique())
print("Route to Market :",sales_data['Route To Market'].unique())
print("Opprtunity Result:",sales_data['Opportunity Result'].unique())
print("Competitor Type : ",sales_data['Competitor Type'].unique())
print("Supplies Group :",sales_data['Supplies Group'].unique())
```

```python
#All these strings will be encoded in the below cell
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
```

```python
sales_data['Supplies Group'] = le.fit_transform(sales_data['Supplies Group'])
sales_data['Region'] = le.fit_transform(sales_data['Region'])
sales_data['Route To Market'] = le.fit_transform(sales_data['Route To Market'])
sales_data['Opportunity Result'] = le.fit_transform(sales_data['Opportunity Result'])
sales_data['Competitor Type'] = le.fit_transform(sales_data['Competitor Type'])
sales_data['Supplies Subgroup'] = le.fit_transform(sales_data['Supplies Subgroup'])
```

```python
sales_data.head()
```

```python
## Now all our data is in the integer format
```

## And now, we will be splitting the data into training and testing sets

```python
cols = [col for col in sales_data.columns if col not in ['Opportunity Number','Opportunity Result']]
```

```python
data = sales_data[cols]
```

```python
data.head()
```

```python
#We've removed the Opportunity Number and OPP result columns as they arent needed
```

```python
# The column 'Opportunity Result' is the target so we store it in  a target variable
```

```python
target = sales_data['Opportunity Result']
```

```python
# For this data,we'll split the dataset into 70:30 ratio for training and testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Splitting data
```

```python
data_train,data_test,target_train,target_test = train_test_split(data,target,test_size=0.30,random_state=10)
```

## Gaussian Naive Bayes 



```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gnb = GaussianNB()

pred = gnb.fit(data_train,target_train).predict(data_test)

print("Naive Bayes accuracy : ",accuracy_score(target_test,pred,normalize=True))
```

## Linear Support Vector Classification

```python
from sklearn.svm import LinearSVC
svc_model = LinearSVC(random_state=0)
pred = svc_model.fit(data_train,target_train).predict(data_test)
print("LinearSVC : ",accuracy_score(target_test,pred,normalize=True))
```

## K-Neighbours Classification

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=3)

pred = neigh.fit(data_train,target_train).predict(data_test)

print("KNeighboursClassifier : ",accuracy_score(target_test,pred,normalize=True))
```

### So, the best prediction we have got is from K-Neighbors Classifier of 0.8145505809979494
